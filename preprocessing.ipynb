```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'weatherAUS.csv'

df = pd.read_csv(file_path)
```

DATA CLEANING

```python
df.duplicated().sum()
```

```python
# Hapus kolom dengan missing value > 40%
df = df.drop(columns=['Evaporation', 'Sunshine', 'Cloud9am', 'Cloud3pm'])
```

```python
# Isi missing numerik dengan median
num_cols = ['MinTemp', 'MaxTemp', 'Rainfall', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
            'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Temp9am', 'Temp3pm']

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Isi missing kategorikal dengan modus
cat_cols = ['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

ENCODING

```python
# Ubah Date ke datetime dan ekstrak fitur waktu
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek
```

```python
df.drop(columns=['Date'], inplace=True)
```

```python
df = pd.get_dummies(df, columns=['Location'], drop_first=True)
location_cols = [col for col in df.columns if col.startswith('Location_')]
df[location_cols] = df[location_cols].astype('int8')
```

```python
df["RainToday"] = df["RainToday"].map({"No": 0, "Yes": 1})
df["RainTomorrow"] = df["RainTomorrow"].map({"No": 0, "Yes": 1})
```

```python
from sklearn.preprocessing import OneHotEncoder

# Mapping arah angin ke derajat
wind_direction_map = {
    "N": 0, "NNE": 22.5, "NE": 45, "ENE": 67.5,
    "E": 90, "ESE": 112.5, "SE": 135, "SSE": 157.5,
    "S": 180, "SSW": 202.5, "SW": 225, "WSW": 247.5,
    "W": 270, "WNW": 292.5, "NW": 315, "NNW": 337.5
}

# Fungsi encoding siklis
def encode_cyclic_feature(df, column):
    df[column + "_deg"] = df[column].map(wind_direction_map)  # Konversi ke derajat
    df[column + "_sin"] = np.sin(np.deg2rad(df[column + "_deg"]))
    df[column + "_cos"] = np.cos(np.deg2rad(df[column + "_deg"]))
    df.drop(columns=[column, column + "_deg"], inplace=True)  # Hapus kolom asli
    return df

# Terapkan encoding ke semua kolom arah angin
for col in ["WindGustDir", "WindDir9am", "WindDir3pm"]:
    df = encode_cyclic_feature(df, col)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, accuracy_score
import joblib
```

```python
# Membagi data untuk prediksi cuaca (RainToday)
X_rain = df[['Rainfall', 'Humidity9am', 'Humidity3pm', 'RainTomorrow']]
y_rain = df['RainToday']

X_train_rain, X_test_rain, y_train_rain, y_test_rain = train_test_split(X_rain, y_rain, test_size=0.2, random_state=42)

# Membuat model RandomForestClassifier untuk prediksi cuaca
model_rain = RandomForestClassifier(n_estimators=100, random_state=42)
model_rain.fit(X_train_rain, y_train_rain)

# Evaluasi model
y_pred_rain = model_rain.predict(X_test_rain)
accuracy_rain = accuracy_score(y_test_rain, y_pred_rain)
print(f'Accuracy for RainToday Prediction: {accuracy_rain}')

# Menyimpan model cuaca
joblib.dump(model_rain, 'model cuaca.pkl')
```

```python
# Membagi data untuk prediksi suhu (MaxTemp)
X_temp = df[['MinTemp', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Temp9am', 'Temp3pm', 'WindDir9am_sin']]
y_temp = df['MaxTemp']

X_train_temp, X_test_temp, y_train_temp, y_test_temp = train_test_split(X_temp, y_temp, test_size=0.2, random_state=42)

# Membuat model RandomForestRegressor untuk prediksi suhu
model_temp = RandomForestRegressor(n_estimators=100, random_state=42)
model_temp.fit(X_train_temp, y_train_temp)

# Evaluasi model
y_pred_temp = model_temp.predict(X_test_temp)
mse_temp = mean_squared_error(y_test_temp, y_pred_temp)
print(f'Mean Squared Error for MaxTemp Prediction: {mse_temp}')

# Menyimpan model suhu
joblib.dump(model_temp, 'model suhu.pkl')
```

```python
# Membagi data untuk prediksi tekanan (Pressure3pm)
X_pressure = df[['MinTemp', 'MaxTemp', 'WindGustSpeed', 'Pressure9am', 'Temp9am', 'Temp3pm']]
y_pressure = df['Pressure3pm']

X_train_pressure, X_test_pressure, y_train_pressure, y_test_pressure = train_test_split(X_pressure, y_pressure, test_size=0.2, random_state=42)

# Membuat model RandomForestRegressor untuk prediksi tekanan
model_pressure = RandomForestRegressor(n_estimators=100, random_state=42)
model_pressure.fit(X_train_pressure, y_train_pressure)

# Evaluasi model
y_pred_pressure = model_pressure.predict(X_test_pressure)
mse_pressure = mean_squared_error(y_test_pressure, y_pred_pressure)
print(f'Mean Squared Error for Pressure3pm Prediction: {mse_pressure}')

# Menyimpan model tekanan
joblib.dump(model_pressure, 'model tekanan.pkl')
```

```python
# Membagi data untuk prediksi kelembaban (Humidity9am)
X_humidity = df[['MaxTemp', 'WindSpeed9am', 'Humidity3pm', 'Temp9am', 'Temp3pm', 'RainToday', 'RainTomorrow']]
y_humidity = df['Humidity9am']

X_train_humidity, X_test_humidity, y_train_humidity, y_test_humidity = train_test_split(X_humidity, y_humidity, test_size=0.2, random_state=42)

# Membuat model RandomForestRegressor untuk prediksi kelembaban
model_humidity = RandomForestRegressor(n_estimators=100, random_state=42)
model_humidity.fit(X_train_humidity, y_train_humidity)

# Evaluasi model
y_pred_humidity = model_humidity.predict(X_test_humidity)
mse_humidity = mean_squared_error(y_test_humidity, y_pred_humidity)
print(f'Mean Squared Error for Humidity9am Prediction: {mse_humidity}')

# Menyimpan model kelembaban
joblib.dump(model_humidity, 'model kelembaban.pkl')
```
